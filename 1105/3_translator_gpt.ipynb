```python
from dotenv import load_dotenv
import openai

# .env í™˜ê²½ë³€ìˆ˜ íŒŒì¼ ê°€ì ¸ì˜¤ê¸°
load_dotenv()
```

```python
client = openai.OpenAI()
```

```python
def translate_text_gpt(text: str, src_lang: str, trg_lang: str) -> str:
    system_instruction = f"""
    ë„ˆëŠ” ë²ˆì—­ ì•±ì´ë‹¤. ì…ë ¥ëœ ë¬¸ì¥ {text}ì„ {src_lang}ì—ì„œ {trg_lang}ë¡œ ìì—°ìŠ¤ëŸ½ê²Œ ë²ˆì—­í•œë‹¤.
    ë²ˆì—­ëœ í…ìŠ¤íŠ¸ë§Œ ì¶œë ¥í•˜ê³  ì¶”ê°€ ì„¤ëª…ì´ë‚˜ ì›ë¬¸ì„ í¬í•¨í•˜ì§€ ì•ŠëŠ”ë‹¤. ê·¸ë¦¬ê³  , ë‹¤ë¥¸ ì§ˆë¬¸ ë˜ëŠ”
    ë²ˆì—­ê³¼ ìƒê´€ì—†ëŠ” ë‚´ìš©ì€ 'ëª¨ë¦…ë‹ˆë‹¤.' ë¡œ ë°˜ë“œì‹œ ë‹µë³€í•˜ê¸°.
    """
    messages = [
        {"role": "system", "content": system_instruction},
        {"role": "user", "content": text},
    ]
    response = client.chat.completions.create(
        model="gpt-4.1-mini",  # ìµœì‹ /ê°€ì„±ë¹„ ëª¨ë¸
        messages=messages,
        temperature=0.2
    )
    return response.choices[0].message.content or ""
    # or ê¸°ì¤€ìœ¼ë¡œ ì•ì˜ ê°’ì´ ì—†ì„ ë•ŒëŠ” ì˜¤ë¥˜ ëŒ€ì‹  "" ë¹ˆ ë¬¸ìì—´(ê¸°ë³¸ê°’) ì²˜ë¦¬
```

```python
text = input('ë²ˆì—­í•  ë‚´ìš©ì„ ì…ë ¥í•˜ê¸°')
src_lang = "í•œêµ­ì–´"
trg_lang = "ì˜ì–´"
translate_text_gpt(text,src_lang,trg_lang)
```

```python
# ì´ í•¨ìˆ˜ëŠ” translate_text_gpt í•¨ìˆ˜ë¥¼ ì‹¤í–‰í•˜ê¸° ì „ì— ìœ íš¨ì„± ê²€ì‚¬í•˜ëŠ” ëª©ì 
def gradio_translate(text: str, src_lang: str, trg_lang: str) -> str:
    if not text.strip():
        return "ë²ˆì—­í•  ë‚´ìš©ì„ ì…ë ¥í•˜ì„¸ìš”."
    if src_lang == trg_lang:
        return "ì›ë³¸ì–¸ì–´ì™€ ë²ˆì—­ì–¸ì–´ë¥¼ ë‹¤ë¥´ê²Œ ì„ íƒí•˜ì„¸ìš”."
    return translate_text_gpt(text,src_lang,trg_lang)
```

```python
import gradio as gr

with gr.Blocks() as demo:
    gr.Markdown('# ğŸ§™â€â™‚ï¸ì´ˆê°„ë‹¨ ë²ˆì—­')
    text_input = gr.Textbox(lines=5, label="ë²ˆì—­í•  ë‚´ìš©ì„ ì…ë ¥í•˜ì„¸ìš”.", placeholder="ì—¬ê¸°ì— ë‚´ìš©ì„ ì…ë ¥í•˜ì„¸ìš”.")
    src_lang_dropdown = gr.Dropdown(choices=["í•œêµ­ì–´","ì˜ì–´","ì¼ë³¸ì–´"], value="í•œêµ­ì–´", label="ì›ë³¸ ì–¸ì–´ë¥¼ ì„ íƒí•˜ì„¸ìš”.")
    trg_lang_dropdown = gr.Dropdown(choices=["í•œêµ­ì–´","ì˜ì–´","ì¼ë³¸ì–´"], value="ì˜ì–´", label="ë²ˆì—­í•  ì–¸ì–´ë¥¼ ì„ íƒí•˜ì„¸ìš”.")
    translate_button = gr.Button("ë²ˆì—­")
    output_text = gr.Textbox(lines=5,label="ë²ˆì—­ ê²°ê³¼",placeholder="ë³€ì—­ëœ ê²°ê³¼ê°€ ì—¬ê¸°ì— í‘œì‹œë©ë‹ˆë‹¤.", interactive=False)

    # ë²„íŠ¼ì´ í´ë¦­ë˜ë©´ ì‹¤í–‰ í•¨ìˆ˜, í•¨ìˆ˜ì˜ ì…ë ¥ê°’, ë¦¬í„´ê²°ê³¼
    translate_button.click(
        fn=gradio_translate,
        inputs=[text_input, src_lang_dropdown, trg_lang_dropdown],
        outputs=output_text
    )
demo.launch(inline=False)
```
