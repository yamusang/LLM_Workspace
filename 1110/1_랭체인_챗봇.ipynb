```python
from langchain_core.messages import SystemMessage, AIMessage
```

```python
messages = [
  SystemMessage("너는 사용자를 도와주는 상담사야.")
]
```

```python
isPrint=True
while True:
    user_input = input("사용자: ")  # 1. 사용자 입력 받기

    if user_input == "exit":  #2.사용자가 대화를 종료하려는지 확인인
        break
    # 랭체인의 메시지 기록(히스토리) 기능
    messages.append(
        HumanMessage(user_input)
    )  # 사용자 메시지를 대화 기록에 추가 
    
    ai_response = llm.invoke(messages)  # 대화 기록을 기반으로 AI 응답 가져오기
    if isPrint:
      print(f"log ai_response ➡ {ai_response.model_dump_json(indent=2)}")
      isPrint=False
    messages.append(
        ai_response
    )  # AI 응답 대화 기록에 추가하기

    print("AI: " + ai_response.content)  # AI 응답 출력
```

```python
print(f'{message.__class_getitem__}\n' for message in messages)
```
